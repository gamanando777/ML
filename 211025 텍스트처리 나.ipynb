* 불용어 제거

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
```

```python
example = "Family is not an important thing. It's everything."

```

```python
# 자주 출현하지만 불필요하다고 생각하는 단어들
# 처리하는 단위 : 단어(어절)

stop_words = set(stopwords.words('english')) 
```

```python
stop_words
```

```python
#주어진 텍스트(문장,문단, 여러 페이지) 쪼개야 한다!
word_tokens = word_tokenize(example)
word_tokens
```

```python

result = []
for w in word_tokens: # 주어진 텍스트를 구분단위로 짤라냄
    if w not in stop_words: # 개별 덩어리가 지워야 할 리스트에 있는지 없는지
        result.append(w) 

print(word_tokens) 
print(result)
```

* wordnet

```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python

# 주어진 처리한 텍스트 데이터
text = "studies studying cries cry"
# 기본적인 덩어리 단위로 쪼개는 것 -> word -> tokenize
tokenization = nltk.word_tokenize(text)
print(tokenization)
print("----")

# 의미 중심으로 처리하는 모델
wordnet_lemmatizer = WordNetLemmatizer()

for w in tokenization:
    print("Lemma for {} is {}".format(w, wordnet_lemmatizer.lemmatize(w)))
```

* n-gram

```python
from nltk import ngrams
```

```python

text = "The increasingly shallow depth meant that, for Tokyo, South Korea brought back a handful of athletes from Rio. Cho, An Baul and An Chang-rim delivered medals, but it doesn't bode well for the long-term future of Korean judo when none of them faced serious challenges in the Olympic trials."
bi_grams = ngrams(text.split(),2)
for g in bi_grams:
    print(g, end=" ")
    
print("\n-----------------------------------------------\n")
tri_grams = ngrams(text.split(),3)
for g in tri_grams:
    print(g, end=" ")
```
